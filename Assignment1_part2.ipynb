```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("clean_chicago_traffic.csv")
```

```python
print(df.columns)
```

```python
df['Police Notified Date'] = pd.to_datetime(df['Police Notified Date'])
df['Crash Month'] = df['Police Notified Date'].dt.to_period('M')


monthly_counts = df['Crash Month'].value_counts().sort_index()

# Plot a line graph
plt.figure(figsize=(12, 8))
monthly_counts.plot(kind='line', marker='o', color='g')
plt.title('Yearly Trend of Accidents')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Plot a scatter plot for geographical distribution
plt.figure(figsize=(10, 8))
sns.scatterplot(x='LONGITUDE', y='LATITUDE', data=df, hue='REPORT_TYPE', palette='tab10', alpha=0.6)
plt.title('Geographical Distribution of Accidents')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.ylim([41, 43])
plt.xlim([-88, -87])
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Move legend outside the plot
plt.tight_layout()
plt.show()
```

```python
location_counts = df['DEVICE_CONDITION'].value_counts()

# Plot a bar graph
plt.figure(figsize=(6, 4))  # Adjust figure size for horizontal bars
sns.barplot(x=location_counts.values, y=location_counts.index, orient='h')
plt.title('Frequency of Accidents')
plt.xlabel('Number of Accidents')
plt.ylabel('Device Condition')
plt.tight_layout()
plt.show()
```

```python
offence_counts = df['POSTED_SPEED_LIMIT'].value_counts()

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(offence_counts, labels=offence_counts.index, autopct='%2.2f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of speed limit')
plt.tight_layout()
plt.show()
```

```python
offence_counts = df['STREET_DIRECTION'].value_counts()

# Plot a pie chart
plt.figure(figsize=(7, 7))
plt.pie(offence_counts, labels=offence_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Distribution of Street Direction')
plt.tight_layout()
plt.show()
```
